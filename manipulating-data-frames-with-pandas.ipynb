### INDEXING DATA FRAMES
* Indexing using square brackets
* Using column attribute and row label
* Using loc accessor
* Selecting only some columns

```python
#read data
data = pd.read_csv('/kaggle/input/pokemon-challenge/pokemon.csv')
data = data.set_index("#") #indexi 1 ile başlatıyor
data.head()
```

```python
data["HP"][1] #indexing using square brackets
```

```python
data.HP[1] #using columns attribute and row label
```

```python
#using loc accessor
data.loc[1,["HP"]] #1.satır HP sütunu
```

```python
#selecting only some columns
data[["HP","Attack"]]
```

### SLICING DATA FRAME
* Difference between selecting columns
  * Series and data frames.
* Slicing and indexing series
* Reverse slicing
* From something to end

```python
#difference between selecting columns: series and dataframes
print(type(data["HP"])) #series
print(type(data[["HP"]])) #dataframe
```

```python
#slicing and indexing series
data.loc[1:10,"HP":"Defense"] #1 to 10 "Defense" between "HP"
```

```python
#reverse slicing 
data.loc[10:1:-1,"HP":"Defense"]
```

```python
#from something to end
data.loc[1:10,"Speed":] #from speed to end.
```

### FILTERING DATA FRAMES 
Creating boolean series Combining filters filtering column based others.

```python
#creating boolean series
boolean = data.HP > 200
data[boolean]
```

```python
#combining filters 
first_filter = data.HP > 150
second_filter = data.Speed > 35
data[first_filter & second_filter]
```

```python
#filtering column based others
data.HP[data.Speed<15] #hızı 15 'ten küçük olan pokemonların canını gösterir.
```

### TRANSFORMING DATA
* Plain python functions
* Lambda function: to apply arbitrary python function to every element

```python
 #plain python func
def div(n):
    return n/2
data.HP.apply(div)#datanın hp satırındakileri alıp 2ye bölüyor.
```

```python
#we can use lambda function
data.HP.apply(lambda n: n/2)
```

```python
#Defining column using other columns
data["total power"] = data.Attack + data.Defense
data.head()
```

### PIVOTING DATA FRAMES
* pivoting:reshape tool

```python
dic = {"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,6,56,8]}
df = pd.DataFrame(dic)
df
```

```python
#pivoting
df.pivot(index = "treatment",columns="gender",values="response")
```

### STACKING and UNSTACKING DATAFRAME
* deal with multi label indexes
* level:position of unstacked index
* swaplevel:change inner and outer level index position

```python
df1 = df.set_index(["treatment","gender"])
df1
```

```python
#level determines indexes
df1.unstack(level=0)
```

```python
df1.unstack(level=1)
```

```python
df2 = df1.swaplevel(0,1)
df2
```

### MELTING DATA FRAMES
* Reverse of pivoting

```python
pd.melt(df,id_vars="treatment",value_vars=["age","response"])
```
