<h3/> Matplotlib <h3/>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/pokemon-challenge/pokemon.csv')
```

```python
data.head()
```

```python
data.Speed.plot(kind = 'line', color = 'g', label = 'Speed', linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Defense.plot(color = 'r',label='Defense',linewidth=1, alpha = 0.5, grid= True,linestyle = '-.')
plt.legend(loc = 'upper right')
plt.xlabel('x axis') # legend = puts label into plot
plt.ylabel('y axis') # label = name of lable
plt.title('Line Plot') #title = title of plot
plt.show()
```

```python
data.plot(kind='scatter',x='Attack',y='Defense',alpha = 0.5, color = 'red')#x,y column name
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack Defense Scatter Plot')
```

```python
#Histogram
#bins = number of bar in figure
data.Speed.plot(kind = 'hist',bins = 50,figsize = (15,15))
plt.show()
```

```python
data.Speed.plot(kind = 'hist',bins = 50)
plt.clf()
```

```python
#keys have to be immutable objects like string, boolean, float, integer or tubles
#list is not immutable
#keys are unique
dictionary['spain'] = "barcelona" #update existing entry
print(dictionary)
dictionary['france'] = 'paris' #add new entry
print(dictionary)
del dictionary['spain'] #remove entry with key 'spain'
print(dictionary)
print('france' in dictionary) #check include or not
dictionary.clear() #remove all entries in dictionary
print(dictionary)
```

<h3/> PANDAS <h3/>


```python
data = pd.read_csv('/kaggle/input/pokemon-challenge/pokemon.csv')
```

```python
series = data['Defense'] #data['Defense'] = series
print(type(series))
data_frame = data[['Defense']] #data[['Defense']] = dataFrame
print(type(data_frame))
```

**logic, control flow, filtering**
<br>Comparison operator: ==, <, >,<=
<br>Boolean operators: and,or,not
<br>Filtering pandas

```python
#Comparison operator
print(3 > 2)
print(3!=2)
#Boolean operators
print(True and False)
print(True or False)
```

```python
# 1- Filtering Pandas data frame
x = data['Defense']>200 #There are only 3 pokemons who have higher defense value than 200
data[x]
```

```python
#2 - Filtering pandas with Logical_and
data[np.logical_and(data['Defense']>200, data['Attack']>100)]
```

```python
#this is alsa same with previous code line.
data[(data['Defense']>200) & (data['Attack']>100)]
```

<h3/> WHILE AND FOR LOOPS <h3/>


```python
 #stay in loop if condition(i is not  equal 5) is true
i = 0
while i!= 5:
    print('i is: ', i)
    i+=1
print(i, 'is equal to 5')
```

```python
lis = [1,2,3,4,5]
for i in lis:
    print('i is: ',i)
print('')
#enumerate index and value of list
#index : value =  0:1, 1:2 , 2:3, 3:4 ,4:5
for index, value in enumerate(lis):
    print(index, " : ",value)
print('')

#For dictionaries 
#We can use for loop to achive key and value of dictionary.
dictionary = {'spain':'madrid', 'france':'paris'}
for key,value in dictionary.items():
    print(key, " : ",value)
print('')

#for pandas we can achieve index and value
for index,value in data[['Attack']][0:1].iterrows():
    print(index," : ",value)
    
```
