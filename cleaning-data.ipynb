### TIDY DATA
We tidy data with melt().

```python
#firtsly ı create new data from pokemons data to explain melt nore easily.
data_new = data.head() #ı only take 5 rows into new data
data_new
```

```python
#Lets melt
#id_vars = what we do not wish to melt
#value_vars = what we want to melt
melted = pd.melt(frame=data_new,id_vars= 'Name',value_vars=['Attack','Defense'])
melted
```

### PIVOTING DATA
Reverse of melting.

```python
# Index is name
# I want to make that columns are variable
# Finally values in columns are value
melted.pivot(index = 'Name',columns = 'variable',values='value')
```

### CONCATENATING DATA
We can concatenate two dataframe.


```python
# Firstly lets create 2 data frame
data1 = data.head()
data2 = data.tail()
conc_data_row = pd.concat([data1,data2],axis = 0,ignore_index=True) #axis = 1:adds dataframes in rows
conc_data_row
```

```python
data1 = data['Attack'].head()
data2 = data['Defense'].head()
conc_data_col = pd.concat([data1,data2],axis=1) #axis=1:adds dataframes in column
conc_data_col
```

### MISSING DATA and TESTING WITH ASSERT
If we encounter with missing data, what we can do:
* leave as is
* drop them with dropna()
* fill missing value wiht fillna()

```python
data.info()
```
